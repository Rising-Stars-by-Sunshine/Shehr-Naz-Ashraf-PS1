## Visualizations

```python
# Pivot data for stacked bar chart - regions
urb_rural_pivot_df = df.pivot_table(index='region', columns='urb_rural', values='PermNum', aggfunc='count')

# Plot stacked bar chart
urb_rural_pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6))

# Formatting
plt.title("Urban vs. Rural Population by Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.legend(title="Urban/Rural")
plt.show()
```

```python
# Add TOTAL column and row (sum modifications across all regions)
urb_rural_pivot_df["TOTAL"] = urb_rural_pivot_df.sum(axis=1)
urb_rural_pivot_df.loc["TOTAL"] = urb_rural_pivot_df.sum()

# print Home totals for regions
print(urb_rural_pivot_df)
```

```python
# Pivot data for stacked bar chart - states
state_urb_rural_pivot_df = df.pivot_table(index='assigned_state', columns='urb_rural', values='PermNum', aggfunc='count')

# Plot stacked bar chart
state_urb_rural_pivot_df.plot(kind='bar', stacked=True, figsize=(12, 6))

# Formatting
plt.title("Urban vs. Rural Population by States Assigned")
plt.xlabel("assigned_state")
plt.ylabel("Count")
plt.legend(title="Urban/Rural")
plt.show()
```

```python
# Add TOTAL column and row (sum modifications across all states)
state_urb_rural_pivot_df["TOTAL"] = state_urb_rural_pivot_df.sum(axis=1)
state_urb_rural_pivot_df.loc["TOTAL"] = state_urb_rural_pivot_df.sum()
```

```python
# Pivot data for stacked bar chart
pivot_df = df.pivot_table(index='assigned_state', columns='hometype', values='PermNum', aggfunc='count')

# Plot stacked bar chart
pivot_df.plot(kind='bar', stacked=True, figsize=(12, 6))

# Formatting
plt.title("Population of Home Types by State Assigned")
plt.xlabel("assigned_state")
plt.ylabel("Count")
plt.legend(title="Home Types")
plt.show()
```

```python
# Pivot data for stacked bar chart
pivot_df = df.pivot_table(index='assigned_state', columns='own_rent', values='PermNum', aggfunc='count')

# Plot stacked bar chart
pivot_df.plot(kind='bar', stacked=True, figsize=(12, 6))

# Formatting
plt.title("Home Ownership for State Assigned")
plt.xlabel("assigned_state")
plt.ylabel("Count")
plt.legend(title="Home Ownership")
plt.show()
```
